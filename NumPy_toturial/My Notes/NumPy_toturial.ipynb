```python
#6. Array Stacking and splitting
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.vstack((a,b))) # vertical stacking - row wise
print(np.hstack((a,b))) # horizontal stacking - column wise
```

```python
c = np.array([[1,2,3], [4,5,6]])
print(c)
```

```python
split = np.hsplit(c,3)  #horizantol split

for s in split:
  print(s)
```

```python
vsplit = np.vsplit(c,2) # vertical split

for s in vsplit:
  print(s)
```

```python
# 1. Create 5 *5 matrix with  values from 1 to 25
# extract last row, firt column, mub matrix containing 3 * 3

import numpy as np

matrix = np.arange(1, 26).reshape(5, 5)
print("Matrix:\n", matrix)
```

```python
# Extract the last row

last_row = matrix[-1]
print("Last Row:", last_row)
```

```python
# Extract the first column

first_column = matrix[:, 0]
print("First Column:", first_column)
```

```python
# Extract the 33 sub-matrix

sub_matrix = matrix[0:3, 0:3]
print("3x3 Sub-matrix:\n", sub_matrix)
```

```python
# 2. Broadcasting :- Multiplying shape- (3,1) matrix with shape - (3,1 )

import numpy as np

a = np.array([[1],
              [2],
              [3]])   # shape (3,1)

b = np.array([[10],
              [20],
              [30]])  # shape (3,1)

result = a * b
print(result)
```
